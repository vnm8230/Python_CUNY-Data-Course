```python
# What is the comparison of the number of schools with more than 100 students scoring 80 or more in Brooklyn vs those in the Bronx
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Import Data File
df = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv") 
```

```python
df.head(10)
```

```python
# Data Column List
df.columns
```

```python
df.isnull().any()
```

```python
#Clean data fron s that is listed 
```

```python
df = df[df['Number Scoring 80 or Above'] != 's']
```

```python
#convert str to int
```

```python
df['Number Scoring 80 or Above'] = df['Number Scoring 80 or Above'].astype(int)
```

```python
nonzero_df = df[df['Number Scoring 80 or Above'] != 0]
```

```python
df_subset = nonzero_df[['School DBN', 'School Name', 'Total Tested', 'Number Scoring 80 or Above']]
```

```python
# create a filtered of with just Brooklyn
```

```python
BKFiltered_df = df_subset[df_subset['School DBN'].str.contains('K', na=False)]
```

```python
BKFiltered_df.head(20)
```

```python
# of schools in BK w more than 100 that passed with an 80 or above
```

```python
SchoolCount=BKFiltered_df[BKFiltered_df['Number Scoring 80 or Above'] >100].groupby('School Name').sum()
```

```python
SchoolCount.plot(kind='bar', color = 'blue')
plt.ylabel('# Tested')
plt.title('Total Tested vs # Scoring 80 +')
plt.show()
```

```python
print(SchoolCount.ngroups)
```

```python
# Create filter for Bronx
```

```python
BXFiltered_df = df_subset[df_subset['School DBN'].str.contains('X', na=False)]
```

```python
BXFiltered_df.head(20)
```

```python
# of schools in BX w more than 100 that passed with an 80 or above
```

```python
SchoolCount_2=BXFiltered_df[BXFiltered_df['Number Scoring 80 or Above'] >100].groupby('School Name').sum()
```

```python
SchoolCount_2.plot(kind='bar', color = 'purple')
plt.ylabel('# Tested')
plt.title('Total Tested vs # Scoring 80 +')
plt.show()
```

```python
print(SchoolCount_2.ngroups)
```

```python
#The analysis indicates that Brooklyn has a significantly higher number of schools with more than 100 students scoring 80 or above on the Regents examinations. Brooklyn has 44 schools meeting this criterion, compared to only 9 schools in the Bronx.
```

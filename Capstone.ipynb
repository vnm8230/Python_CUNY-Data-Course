```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import the dataset that will be used for analysis
```

```python
df_crash = pd.read_csv('Motor_Vehicle_Collisions_-_Person_20260427.csv')
```

```python
df_crash.isnull().any()
```

```python
# Create a subset of the dataset with specific columns
```

```python
df_crashsubset = df_crash[['CRASH_DATE', 'PERSON_TYPE', 'PED_ROLE', 'PERSON_AGE', 'BODILY_INJURY', 'PERSON_SEX']]
```

```python
# Check column creation
```

```python
df_crashsubset.columns
```

```python
df_crashsubset.head(10)
```

```python
# filter out NaN data from the dataset
```

```python
df_crashsubset = df_crashsubset.dropna(subset=["PERSON_AGE", "BODILY_INJURY", "PERSON_SEX","PED_ROLE"])
```

```python
#Check for filter
```

```python
df_crashsubset.head(10)
```

```python
# Need to convert PERSON AGE from a string 
```

```python
df_crashsubset['PERSON_AGE'] = pd.to_numeric(df_crashsubset['PERSON_AGE'], errors='coerce')
```

```python
# Convert CRASH_DATE to a datetime
```

```python
df_crashsubset['CRASH_DATE'] = pd.to_datetime(df_crashsubset['CRASH_DATE'])
```

```python
# Filter out Data to compare and narrow the age group and date. 
```

```python
df_crashhead_Male = df_crashsubset[
    (df_crashsubset['CRASH_DATE'].between('2020-01-01', '2026-04-30')) &
    (df_crashsubset['PERSON_SEX'] == 'M') &
    (df_crashsubset['BODILY_INJURY'] == 'Head') &
    (df_crashsubset['PERSON_TYPE'] == 'Occupant') &
    (df_crashsubset['PED_ROLE'] == 'Driver') &
    (df_crashsubset['PERSON_AGE'] > 16) &
    (df_crashsubset['PERSON_AGE'] < 30)]
```

```python
df_crashhead_Male.head(10)
```

```python
#Count the # of rows in the filtered data set for head injury
```

```python
row_count_Head_M = df_crashhead_Male.shape[0]
print(row_count_Head_M)
```

```python
# Filter Data set for no bodily injury crashes
```

```python
df_crashDNA_Male = df_crashsubset[(df_crashsubset['CRASH_DATE'].between('2020-01-01', '2026-04-30')) & (df_crashsubset['PERSON_SEX'] == 'M') & (df_crashsubset['BODILY_INJURY'] == 'Does Not Apply') & (df_crashsubset['PERSON_TYPE'] == 'Occupant') & (df_crashsubset['PED_ROLE'] == 'Driver')]
```

```python
df_crashDNA_Male.head(10)
```

```python
#Count the # of rows in the filtered data set for head injury
```

```python
row_count_DNA_M = df_crashDNA_Male.shape[0]
print(row_count_DNA_M)
```

```python
# Total number of rows in both data sets
```

```python
Men_TotalAccidents = (row_count_Head_M + row_count_DNA_M)
```

```python
print(Men_TotalAccidents)
```

```python
# Create a visualization
```

```python
CountMen_head = len(df_crashhead_Male)
CountMen_DNA = len(df_crashDNA_Male)
```

```python
groups = ['Head Injury', 'No Bodily Injury']
counts = [CountMen_head, CountMen_DNA]
```

```python
plt.ylabel('Number of Accidents (millions)')
plt.bar(groups, counts)
for i, v in enumerate(counts):
    plt.text(i, v, str(v), ha='center', va='bottom')  # Changed from counts.text() to plt.text()
```

```python
# create the same Data sets for women drivers
```

```python
df_crashhead_Female.head(10)
```

```python
df_crashDNA_Female = df_crashsubset[(df_crashsubset['CRASH_DATE'].between('2020-01-01', '2026-04-30')) & (df_crashsubset['PERSON_SEX'] == 'F') & (df_crashsubset['BODILY_INJURY'] == 'Does Not Apply') & (df_crashsubset['PERSON_TYPE'] == 'Occupant') & (df_crashsubset['PED_ROLE'] == 'Driver')]
```

```python
df_crashDNA_Female.head(10)
```

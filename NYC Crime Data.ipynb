```python
# What is the number of Felony Assaults being committed by Caucasians compared to those committed by Hispanics
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('NYPD_Arrest_Data_(Year_to_Date)_20260407.csv')
```

```python
df.head(10)
```

```python
#Clear all null data 
```

```python
df = df[df['AGE_GROUP'] != '(null)']
```

```python
df = df[df['PERP_SEX'] != '(null)']
```

```python
df.head(10)
```

```python
#create a subset with necessary columns
```

```python
df_arrest_subset= df[['ARREST_DATE', 'OFNS_DESC', 'ARREST_BORO', 'AGE_GROUP', 'PERP_SEX', 'PERP_RACE']] 
```

```python
df_arrest_subset.head(20)
```

```python
# Filter for 
```

```python
ManFiltered_df = df_arrest_subset[df_arrest_subset['ARREST_BORO'].str.contains('M', na=False)]
```

```python
Man_sexMen_df = ManFiltered_df[ManFiltered_df['PERP_SEX'].str.contains('M', na=False)]
```

```python
Man_sexMen_df.head(20)
```

```python
Arrestype_df= Man_sexMen_df[Man_sexMen_df['OFNS_DESC'].str.contains('FELONY ASSAULT' , na=False) & Man_sexMen_df['AGE_GROUP'].str.contains('25-44', na=False)]
```

```python
Arrestype_df.head(20)
```

```python
Arrestypewht_df= Arrestype_df[Arrestype_df['PERP_RACE'].str.contains('WHITE', na=False)]
```

```python
Arrestypehis_df= Arrestype_df[Arrestype_df['PERP_RACE'].str.contains('WHITE HISPANIC', na=False)]
```

```python
Countwht = len(Arrestypewht_df)
Counthis = len(Arrestypehis_df)
```

```python
groups = ['White', 'White Hispanic']
counts = [Countwht, Counthis]
```

```python
plt.ylabel('Number of Fellony Assault Arrests')
plt.bar(groups, counts)
```

```python
# The number of assault arrests for Caucasians in Manhattan is higher than that for Hispanics. 
```

```python
BKFiltered_df = df_arrest_subset[df_arrest_subset['ARREST_BORO'].str.contains('K', na=False)]
```

```python
BK_sexMen_df = BKFiltered_df[BKFiltered_df['PERP_SEX'].str.contains('M', na=False)]
```

```python
BK_sexMen_df.head(10)
```

```python
BK_Arrestype_df= BK_sexMen_df[BK_sexMen_df['OFNS_DESC'].str.contains('FELONY ASSAULT' , na=False) & BK_sexMen_df['AGE_GROUP'].str.contains('25-44', na=False)]
```

```python
BK_Arrestypewht_df= BK_Arrestype_df[BK_Arrestype_df['PERP_RACE'].str.contains('WHITE', na=False)]
```

```python
BK_Arrestypehis_df= BK_Arrestype_df[BK_Arrestype_df['PERP_RACE'].str.contains('WHITE HISPANIC', na=False)]
```

```python
BK_countwht= len(BK_Arrestypewht_df)
BK_counthis= len(BK_Arrestypehis_df)
```

```python
BK_groups = ['White', 'White Hispanic']
BK_counts = [BK_countwht, BK_counthis]
```

```python
plt.ylabel('Number of Fellony Assault Arrests Brooklyn')
plt.bar(BK_groups, BK_counts)
```

```python
# The number of assault arrests for Caucasians in Brooklyn is higher than that for Hispanics. Comparing 2 boroughs, Brooklyn and Manhattan, the number of assaults
# arrests for Caucasians are higher than those for Hispanics. 
```
